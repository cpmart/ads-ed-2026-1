# Dicionários em Python

## O que são dicionários?
Dicionários em Python são estruturas de dados que armazenam pares de **chave: valor**. Cada chave é única e está associada a um valor. Dicionários são mutáveis, ou seja, seus elementos podem ser modificados após a criação.

**Sintaxe básica de um dicionário:**

````python
meu_dicionario = {
    "chave1": "valor1",
    "chave2": "valor2",
    "chave3": "valor3"
}
````


## Características dos dicionários:
- **Chaves únicas**: Não podem existir chaves duplicadas.
- **Mutabilidade**: Podemos alterar valores, adicionar ou remover pares.
- **Acesso rápido**: Chaves permitem um acesso eficiente aos valores.

## Exemplos de uso 

### Exemplo básico: Criação e manipulação
**Objetivo**: Criar um dicionário simples e acessar valores por chave.


````python
# Criação de um dicionário
aluno = {
    "nome": "João",
    "idade": 21,
    "curso": "Engenharia Civil"
}

# Acessando um valor por sua chave
print(aluno["nome"])  # Saída: João

# Modificando um valor
aluno["idade"] = 22

# Adicionando um novo par chave: valor
aluno["matrícula"] = "202410101"

# Removendo uma chave
del aluno["curso"]

print(aluno)

# saída: 
````

### Exemplo de uso de dicionários
**Objetivo**: Este exemplo simples demonstra como criar e acessar um dicionário:


````python
# Dicionário de alunos e suas idades
alunos = {
    "Ana": 20,
    "João": 22,
    "Maria": 21
}

# Acessando o valor de uma chave
idade_ana = alunos["Ana"]
print(f"A idade de Ana é: {idade_ana}")  # Saída: A idade de Ana é: 20

# Calculando a média das idades
idades = alunos.values()  # Obtém todas as idades
print(f"{idades}")
media_idades = sum(idades) / len(idades)  # Soma todas as idades e divide pela quantidade de alunos

print(f"A média das idades é: {media_idades:.2f}")

#saída:
````

### Exemplo de dicionário com listas
**Objetivo**: Aqui, cada chave do dicionário armazena uma lista como valor:


````python
# Dicionário de alunos e suas notas em três disciplinas
notas_alunos = {
    "Ana": [8.5, 7.0, 9.0],
    "João": [6.5, 7.5, 8.0],
    "Maria": [9.0, 9.5, 10.0]
}

# Acessando a lista de notas de João
notas_joao = notas_alunos["João"]
print(f"Notas de João: {notas_joao}") 

#saída:
````

### Exemplo de dicionários com listas e laços
**Objetivo**: Este exemplo mais avançado combina dicionários, listas e laços para calcular a média das notas de cada aluno.

````python
# Dicionário com listas de notas
notas_alunos = {
    "Ana": [8.5, 7.0, 9.0],
    "João": [6.5, 7.5, 8.0],
    "Maria": [9.0, 9.5, 10.0]
}

# Iterando pelo dicionário para calcular e exibir a média de cada aluno
for aluno, notas in notas_alunos.items():
    media = sum(notas) / len(notas)  # Calcula a média
    print(f"Média de {aluno}: {media:.2f}")

# saída:
````

### Exemplo intermediário: Iterando e verificando chaves/valores
**Objetivo**: Percorrer um dicionário, verificar a existência de chaves, e iterar sobre chaves e valores.

````python
produto = {
    "nome": "Laptop",
    "preço": 3500,
    "estoque": 10
}
# Verificando se uma chave existe
if "Preço" in produto:
    print(f"Preço do produto: R$ {produto['preço']}")

# Iterando sobre as chaves
for chave in produto:
    print(f"Chave: {chave}, Valor: {produto[chave]}")

# Ou iterando sobre pares chave: valor diretamente
for chave, valor in produto.items():
    print(f"{chave}: {valor}")

#saída:
````

### Exemplo avançado: Usando dicionários para contar frequências
**Objetivo**: Contar a frequência de palavras em uma frase usando um dicionário.


````python
frase = "python é ótimo porque python é versátil"

# Dividindo a frase em palavras
palavras = frase.split()

print(palavras)
# Contando a frequência das palavras
frequencia = {}
for palavra in palavras:
    if palavra in frequencia:
        frequencia[palavra] += 1
    else:
        frequencia[palavra] = 1

print(frequencia)
# saída:
````


### Exemplo avançado: Aninhamento de dicionários
**Objetivo**: Criar um dicionário de dicionários para organizar informações mais complexas.


````python
empresa = {
    "engenheiros": {
        "João": {"idade": 28, "salário": 8000},
        "Maria": {"idade": 34, "salário": 9500}
    },
    "administrativo": {
        "Carlos": {"idade": 45, "salário": 7000},
        "Ana": {"idade": 50, "salário": 7800}
    }
}
# Acessando informações específicas
print(empresa["engenheiros"]["João"]["salário"])  

# Iterando sobre dicionários aninhados
for setor, funcionarios in empresa.items():
    print(f"Setor: {setor}")
    for nome, detalhes in funcionarios.items():
        print(f"Funcionário: {nome}, Detalhes: {detalhes}")

#saída:
````

## Operações comuns em dicionários:
- **Adicionar item**: `dicionario[chave] = valor`
- **Remover item**: `del dicionario[chave]`
- **Obter valor com fallback**: `dicionario.get(chave, valor_default)`
- **Iterar sobre chaves**: `for chave in dicionario`
- **Iterar sobre valores**: `for valor in dicionario.values()`
- **Iterar sobre pares chave-valor**: `for chave, valor in dicionario.items()`

A linha `if __name__ == "__main__":` é uma parte importante do código Python para controlar a execução de scripts. Vamos explicar essa estrutura em detalhes:

### **Por que ela é usada?**

Em Python, os arquivos `.py` podem ser executados de duas maneiras:
- **Como um script principal**: Quando você executa o arquivo diretamente (por exemplo, `python nome_do_arquivo.py`), o Python o considera o ponto de entrada do programa.
- **Como um módulo**: Quando você importa um arquivo `.py` para outro script (por exemplo, `import nome_do_arquivo`), o Python carrega o conteúdo desse arquivo, mas **não** executa o código fora de funções/classes automaticamente.

Para diferenciar esses dois casos, usamos a estrutura `if __name__ == "__main__":`. Ela verifica se o arquivo está sendo executado diretamente como o script principal e, se estiver, executa o código dentro desse bloco. Caso contrário, ele não o executa, permitindo que o arquivo seja importado sem efeitos colaterais.

### **Como funciona?**

Quando o interpretador Python executa um arquivo, ele atribui o valor `"__main__"` à variável interna `__name__` se o arquivo estiver sendo executado diretamente. Se o arquivo estiver sendo importado como um módulo, o valor de `__name__` será o nome do arquivo ou módulo.

### **Aplicação no seu código**

No seu exemplo:
```python
if __name__ == "__main__":
    main()
```

Isso significa que **se o arquivo for executado diretamente**, a função `main()` será chamada e o programa começará a rodar. Caso o arquivo seja importado para outro script, a função `main()` **não será executada automaticamente**, o que é útil para reutilizar funções ou classes em outros contextos sem iniciar o programa acidentalmente.

### **Por que é importante?**

1. **Reutilização de código**: Permite que o arquivo funcione tanto como script executável quanto como um módulo importável.
2. **Evita efeitos colaterais**: Se você importar o arquivo em outro lugar, a parte principal do código (como o menu da agenda) não será executada automaticamente.
3. **Organização**: Separa o código de execução direta daquele que pode ser reutilizado em outros contextos, como funções e classes.

### Exemplo para demonstrar:

**Arquivo `agenda.py`:**

````python
def exibir_menu():
    print("Exibindo menu...")

if __name__ == "__main__":
    exibir_menu()
````

**Arquivo `outro_programa.py`:**

````python
import agenda

print("Importei o arquivo agenda.py")

#saída:
````

```

- **Ao executar `agenda.py` diretamente:**

Essa separação é essencial para modularizar o código, tornando-o mais flexível e reutilizável.
