## Definição

Uma **lista linear** é uma coleção ordenada de elementos, onde cada elemento ocupa
uma posição específica.

Existem dois tipos principais:

1. **Lista Simples (array/list)**  

   - Os elementos ficam armazenados em **bloco contínuo de memória**.
   - O acesso por índice é rápido (`O(1)`).
   - Inserções/remoções no meio da lista podem ser lentas (`O(n)`).

2. **Lista Encadeada**  

   - Cada elemento (nó) contém um **valor** e um **ponteiro** para o próximo nó.
   - Os nós podem estar espalhados na memória.
   - Inserções/remoções são rápidas no início/fim (`O(1)`), mas o acesso é mais lento (`O(n)`).

## Lista Simples em Python (`list`)

O tipo **list** do Python é dinâmico e pode armazenar diferentes tipos de dados.

#### Operações básicas:
- Inserção (`append`, `insert`)
- Remoção (`remove`, `pop`)
- Busca (`in`, `index`)
- Atualização (`lista[pos] = valor`)
- Exibição (`print(lista)`)


````python
# Criando uma lista simples
lista = [1, 2, 3, 4, 5]
print("Lista original:", lista)

# Inserção no final
lista.append(6)
print("Após inserir 6:", lista)

# Inserção em posição específica
lista.insert(2, 10)
print("Após inserir 10 na posição 2:", lista)

# Remoção de um elemento específico
lista.remove(4)
print("Após remover 4:", lista)

# Atualização de um valor
lista[0] = 100
print("Após atualizar o primeiro valor:", lista)

# Busca por um valor
pos = lista.index(10)
print(f"O valor 10 está na posição: {pos}")

# Exibição da lista final
print("Lista final:", lista)
````

````python
lista = [1, 2, 3, 4, 5]
print(lista)
print(*lista)
````

## Lista Encadeada

Uma lista encadeada é composta por **nós**, onde cada nó contém:

1. Um **valor**.
2. Um **ponteiro** para o próximo nó.

### Estrutura:
```
[valor | próximo] -> [valor | próximo] -> ... -> None
```

### Operações implementadas:

- Inserção no início
- Inserção no fim
- Remoção no início
- Remoção no fim
- Exibição da lista
- Busca por elemento


````python
# Definição de um nó
class No:
    def __init__(self, valor):
        self.valor = valor
        self.proximo = None

# Definição da lista encadeada
class ListaEncadeada:
    def __init__(self):
        self.inicio_lista = None

    # Inserção no início
    def inserir_inicio(self, valor):
        novo_no = No(valor)
        novo_no.proximo = self.inicio_lista
        self.inicio_lista = novo_no

    # Inserção no fim
    def inserir_fim(self, valor):
        novo_no = No(valor)
        if not self.inicio_lista:
            self.inicio_lista = novo_no
        else:
            atual = self.inicio_lista
            while atual.proximo:
                atual = atual.proximo
            atual.proximo = novo_no

    # Remoção no início
    def remover_inicio(self):
        if self.inicio_lista:
            self.inicio_lista = self.inicio_lista.proximo

    # Remoção no fim
    def remover_fim(self):
        if not self.inicio_lista:       #Se lista estiver vazia
            return
        if not self.inicio_lista.proximo: #Se lista estiver com apenas 1
            self.inicio_lista = None
            return
        atual = self.inicio_lista
        while atual.proximo.proximo:
            atual = atual.proximo
        atual.proximo = None

    # Exibir lista
    def exibir(self):
        atual = self.inicio_lista
        while atual:
            print(atual.valor, end=" -> ")
            atual = atual.proximo
        print("None")

    # Buscar elemento
    def buscar(self, valor):
        atual = self.inicio_lista
        pos = 0
        while atual:
            if atual.valor == valor:
                return pos
            atual = atual.proximo
            pos += 1
        return -1

# Testando a lista encadeada
lista_enc = ListaEncadeada()
lista_enc.inserir_inicio(2)
lista_enc.inserir_inicio(1)
lista_enc.inserir_fim(3)
lista_enc.inserir_fim(4)

print("Lista Encadeada:")
lista_enc.exibir()

lista_enc.remover_inicio()
print("Após remover início:")
lista_enc.exibir()

lista_enc.remover_fim()
print("Após remover fim:")
lista_enc.exibir()

print("Buscar valor 2:", lista_enc.buscar(2))
````

## Busca em Listas

### Busca Linear

Percorre a lista elemento por elemento até encontrar o valor desejado.  
Complexidade: **O(n)**

### Busca Binária

- Requer que a lista esteja **ordenada**.  
- Divide a lista ao meio sucessivamente até encontrar o valor.  
- Complexidade: **O(log n)**


````python

# Busca Linear
def busca_linear(lista, alvo):
    for i, v in enumerate(lista):
        if v == alvo:
            return i
    return -1

# Busca Binária
def busca_binaria(lista, alvo):
    inicio, fim = 0, len(lista) - 1
    while inicio <= fim:
        meio = (inicio + fim) // 2
        if lista[meio] == alvo:
            return meio
        elif lista[meio] < alvo:
            inicio = meio + 1
        else:
            fim = meio - 1
    return -1

# Testando as buscas
dados = [1, 3, 5, 7, 9, 11, 13, 15]
print("Lista ordenada:", dados)

print("Busca Linear pelo 7:", busca_linear(dados, 7))
print("Busca Binária pelo 7:", busca_binaria(dados, 7))
````
